```python
# Define the key columns to check for Benin dataset
key_columns = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# 1. Check missing values in key columns before cleaning
print("Missing values in key columns before cleaning:")
print(df[key_columns].isna().sum())

# 2. Compute Z-scores for key columns, ignoring NaNs
z_scores = df[key_columns].apply(zscore, nan_policy='omit')

# 3. Flag rows where any absolute Z-score > 3 (potential outliers)
outliers = (z_scores.abs() > 3)
print(f"\nNumber of outlier rows detected (|Z| > 3) in any key column: {outliers.any(axis=1).sum()}")

# Optionally, get indices of outlier rows
outlier_indices = outliers.any(axis=1)
# print(df.loc[outlier_indices])  # Uncomment to inspect outliers if needed

# 4. Impute missing values in key columns with median values
for col in key_columns:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# 5. Drop rows flagged as outliers (optional)
df_cleaned = df[~outlier_indices].copy()

print(f"\nShape before cleaning: {df.shape}")
print(f"Shape after removing outliers: {df_cleaned.shape}")

# 6. Export cleaned DataFrame to CSV in data folder for Benin
df_cleaned.to_csv('../data/benin_clean.csv', index=False)
print("\nCleaned data exported to '../data/benin_clean.csv'")
```

```python


# Load your dataset (adjust path as needed)
df = pd.read_csv("../data/benin-malanville.csv")

# Define columns to check
columns_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute z-scores for those columns
z_scores = df[columns_to_check].apply(lambda x: (x - x.mean()) / x.std())

# Find rows with any z-score > 3 or < -3 (outliers)
outliers = (np.abs(z_scores) > 3).any(axis=1)

# Add a Cleaning flag: 0 for outliers (pre-clean), 1 for clean data (post-clean)
df['Cleaning'] = 0
df.loc[~outliers, 'Cleaning'] = 1

# Create cleaned DataFrame
df_clean = df[df['Cleaning'] == 1].copy()

# Display counts to verify
print("Rows flagged as outliers:", outliers.sum())
print("Rows kept after cleaning:", len(df_clean))
```

```python


# Assume df is original data, and outliers is a boolean mask of flagged rows
# Create a 'Cleaning' column: 0 = pre-cleaning (all data), 1 = post-cleaning (rows without outliers)
df['Cleaning'] = 0
df.loc[~outliers, 'Cleaning'] = 1

# Group by cleaning flag and calculate average ModA and ModB
grouped = df.groupby('Cleaning')[['ModA', 'ModB']].mean()

print("Average ModA and ModB pre/post-cleaning:")
print(grouped)

# Plot the averages
grouped.plot(kind='bar', figsize=(8,5), title='Average ModA and ModB Pre vs Post Cleaning')
plt.xlabel('Cleaning Flag (0=Pre, 1=Post)')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.show()
```
